Notebook ini berisi ringkasan materi Week 3 yang menjadi fondasi wajib sebelum masuk ke materi pemrosesan citra dengan OpenCV (`cv2`). Di Computer Vision, gambar dibaca dan diproses sebagai **NumPy Array** dan divisualisasikan menggunakan **Matplotlib**.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. NumPy Array & Manipulasinya
Gambar pada OpenCV direpresentasikan sebagai angka (piksel) dalam bentuk matriks (array).
- **Gambar Grayscale** = Array 2D (Dimensi: Tinggi x Lebar)
- **Gambar Berwarna (RGB/BGR)** = Array 3D (Dimensi: Tinggi x Lebar x 3 Channel)

```python
# Membuat 'kanvas kosong' berwarna hitam (0) dan putih (255)
# Sangat berguna untuk membuat Masking (menutupi area tertentu pada citra) di OpenCV

kanvas_hitam = np.zeros((5, 5), dtype=np.uint8)
kanvas_putih = np.ones((5, 5), dtype=np.uint8) * 255

print("Kanvas Hitam:\n", kanvas_hitam)
print("\nKanvas Putih:\n", kanvas_putih)
```

### Indexing & Slicing (Konsep dasar Cropping)
Digunakan untuk memotong gambar (Cropping) atau mengambil *Region of Interest (ROI)*, yaitu fokus area tertentu pada gambar.

```python
# Simulasi gambar 5x5 berisikan nilai piksel random (0-255)
gambar_dummy = np.random.randint(0, 256, (5, 5), dtype=np.uint8)
print("Gambar Dummy Asli:\n", gambar_dummy)

# Slicing: Mengambil piksel dari baris indeks 1 sampai 3, dan kolom indeks 1 sampai 3
crop_gambar = gambar_dummy[1:4, 1:4]
print("\nHasil Crop:\n", crop_gambar)
```

### Stacking (Menggabungkan Array/Gambar)
Sering digunakan untuk menampilkan dua hasil proses gambar (contoh: Gambar Asli vs Gambar Blur) saling bersandingan (kiri-kanan / atas-bawah).

```python
arr1 = np.ones((3, 3), dtype=np.uint8) * 100
arr2 = np.zeros((3, 3), dtype=np.uint8)

# Menggabungkan secara horizontal (kiri-kanan)
gabung_horizontal = np.hstack((arr1, arr2))
print("Gabungan Horizontal:\n", gabung_horizontal)

# Menggabungkan secara vertikal (atas-bawah)
gabung_vertikal = np.vstack((arr1, arr2))
print("\nGabungan Vertikal:\n", gabung_vertikal)
```

## 2. Matplotlib untuk Visualisasi Gambar
Fungsi bawaan OpenCV seperti `cv2.imshow` terkadang bermasalah atau membuat crash di dalam Jupyter Notebook. Matplotlib dengan fungsi `plt.imshow` adalah standar industri untuk memvisualisasikan citra di Notebook.

```python
# Menampilkan array NumPy 2D sebagai gambar abu-abu (grayscale)
plt.figure(figsize=(4,4))
plt.imshow(gambar_dummy, cmap='gray', vmin=0, vmax=255)
plt.title('Visualisasi Matriks Dummy sebagai Grayscale')
plt.colorbar()
plt.show()
```

### Menggunakan Subplot untuk Komparasi
Sangat berguna jika ingin melihat perbandingan langkah-langkah *Computer Vision* (contoh: *Before* vs *After*).

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].imshow(gambar_dummy, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Gambar Asli')

axs[1].imshow(crop_gambar, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Gambar Hasil Crop')

plt.show()
```

## ⚠️ Catatan Penting Sebelum Masuk Week 4 (OpenCV):
1. **Perbedaan Channel Warna**: Matplotlib membaca warna gambar dengan urutan **RGB** (Red-Green-Blue), sedangkan OpenCV membacanya dengan urutan **BGR** (Blue-Green-Red).
2. **Trik Konversi**: Jika kamu membaca gambar pakai `cv2.imread()`, warnanya akan aneh jika langsung di-plot Matplotlib. Gunakan trik Slicing ini untuk membaliknya: 
   `gambar_rgb = gambar_bgr[:, :, ::-1]`
3. **Tipe Data `uint8`**: Piksel gambar berada pada rentang angka **0 - 255**, sehingga tipe data array-nya wajib `np.uint8`. Jika tipenya *float* atau *int32*, Matplotlib dan OpenCV mungkin gagal menampilkan warna dengan benar.
